```python
import pandas as pd
import numpy as np

# Load the dataset
mid_path = "/Users/arkajyotisaha/Desktop/My-Thesis/code/src/data/processed/NASDAQ_ITCH/nasdaq_AAPL_1ms_test.parquet"
df = pd.read_parquet(mid_path, engine="pyarrow")

print("=== NASDAQ ITCH Dataset Analysis ===")
print(f"Shape: {df.shape}")
print(f"Index dtype: {df.index.dtype}")
print(f"Index name: {df.index.name}")

# Check for timestamp columns
print("\n=== Column Analysis ===")
print(f"Total columns: {len(df.columns)}")
timestamp_cols = [col for col in df.columns if 'time' in col.lower()]
label_cols = [col for col in df.columns if 'y' in col.lower() or 'label' in col.lower()]
print(f"Timestamp columns: {timestamp_cols}")
print(f"Label columns: {label_cols}")

# Show data types and missing values
print("\n=== Data Info ===")
print(df.info(show_counts=True))

# Sample the data
print("\n=== First 5 Rows ===")
print(df.head())

print("\n=== Last 5 Rows ===")
print(df.tail())

# Show some statistics
print("\n=== Basic Statistics ===")
if label_cols:
    for label_col in label_cols:
        print(f"\n{label_col} distribution:")
        print(df[label_col].value_counts().sort_index())

# Time range
print(f"\nTime range: {df.index.min()} → {df.index.max()}")
print(f"Duration: {df.index.max() - df.index.min()}")

# Feature columns (excluding labels)
feature_cols = [col for col in df.columns if col not in label_cols]
print(f"\nFeature columns ({len(feature_cols)}): {feature_cols[:10]}...")  # Show first 10

# Check for missing values
missing_counts = df.isnull().sum()
if missing_counts.sum() > 0:
    print(f"\n=== Missing Values ===")
    print(missing_counts[missing_counts > 0])
else:
    print("\n✅ No missing values found")

# Random sample
print("\n=== Random Sample (3 rows) ===")
print(df.sample(3))
```

```python
import pandas as pd, os, datetime as dt

RAW_H5 = "/Users/arkajyotisaha/Desktop/My-Thesis/code/src/data/raw/NASDAQ_ITCH/itch.h5"     # update path if needed

with pd.HDFStore(RAW_H5, "r") as store:
    keys = [k.lstrip("/") for k in store.keys()]
    print("Tables found:", keys)
    for k in keys:
        df = store.select(k, start=0, stop=3)          # peek 1‑st 3 rows
        print(f"\n{k}: total_rows={store.get_storer(k).nrows}")
        if {"timestamp_nanos","seconds","nanoseconds"}.intersection(df.columns):
            # convert first timestamp to human‑readable
            if "timestamp_nanos" in df.columns:
                ts = pd.to_datetime(df["timestamp_nanos"].iloc[0], unit="ns",
                                    utc=True, errors="coerce")
            else:
                ns = df["seconds"].iloc[0]*1_000_000_000 + df["nanoseconds"].iloc[0]
                ts = pd.to_datetime(ns, unit="ns", utc=True, errors="coerce")
            print("   first event:", ts)
        else:
            print(df.size)
```

```python
import os

def print_directory_tree(root_dir, indent=''):
    for item in sorted(os.listdir(root_dir)):
        if item == 'thesis':  # Ignore the "Thesis" folder
            continue
        path = os.path.join(root_dir, item)
        print(indent + '├── ' + item)
        if os.path.isdir(path):
            print_directory_tree(path, indent + '│   ')

print_directory_tree('/Users/arkajyotisaha/Desktop/My-Thesis')  # Use the existing 'root' variable
```
